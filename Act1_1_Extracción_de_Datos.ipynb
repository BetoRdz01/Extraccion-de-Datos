```python
import pandas as pd
```

```python
df = pd.read_csv('Detalle precios y productos fabricados 2022.csv', index_col=0)
df.head()
```

```python
#Filtro por Nombre 
filtro1=df[df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1.to_csv('Filtro1.csv')
filtro1
```

```python
#Filtro por filas
filtro2=df.iloc[ [5, 10, 20, 30, 40, 50] , : ]  #Filas no consecutivas 
filtro2
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , [1,3,4,5,6,7,8]]  #Columnas no consecutivas 
filtro3
```

```python
#Filtro 4: Aplicamos Filtro de filas vs columnas 
filtro4=df.loc[["F37034","F37043"], ["COSTO_UNITARIO_CALCULADO"]]
filtro4
```

```python
#Filtro por cabecera
filtro5=df.head(10)
filtro5
```

```python
#Filtro por comparacion en costo unitario calculado
filtro6=df[df["COSTO_UNITARIO_CALCULADO"] == '$52.80']
filtro6
```

```python
#Filtro de columna cantidad y columna nombre
filtro7=df[(df["CANT"] > 3) & (df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro7
```

```python
#Filtro de columna cantidad o costo unitario
filtro8=df[(df["CANT"] < 10) | (df["COSTO_UNITARIO"] == "$124")]
filtro8
```

```python
#Filtro not con columnas nombres y cantidad
filtro9=df[~(df["NOMBRE"] == 'LETICIA RAMIREZ HERNANDEZ') & ~(df["CANT"] == 1)]
filtro9
```

```python
filtro2.to_csv('Filtro1.csv')
filtro3.to_csv('Filtro2.csv')
filtro4.to_csv('Filtro3.csv')
filtro5.to_csv('Filtro4.csv')
filtro6.to_csv('Filtro5.csv')
filtro7.to_csv('Filtro6.csv')
filtro8.to_csv('Filtro7.csv')
filtro9.to_csv('Filtro8.csv')
```
