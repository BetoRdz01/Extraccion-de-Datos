```python
import pandas as pd
```

```python
df = pd.read_excel('Datos de Facturación.xlsx')
df.head()
```

```python
df.info()
```

```python
#Filtro CVE_CLPV
df['CVE_CLPV'] = df['CVE_CLPV'].replace('[^0-9]+', '0', regex=True)
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)
filtro1_2=df.loc[(df["CVE_CLPV"] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1_2
```

```python
#filtro cve_vend
filtro2_2 = df[~(df['CVE_VEND'] == 4) & ~(df['CVE_VEND'] == 5)]
filtro2_2['CVE_VEND'].value_counts()
```

```python
df.info()
```

```python
df['FECHA_ENT']
```

```python
#Filtro Fecha_ent
# filtro3_2 = df[df['FECHA_ENT'] == '2022-02-28 00:00:00']
filtro3_2 = df[df['FECHA_ENT'] == '2023-03-01 00:00:00']

filtro3_2 # Error con formato de la fecha
```

```python
#filtro de cantitidad total y estatus
filtro4_2 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]
filtro4_2
```

```python
#filtramos columnas
filtro5_2 = df[['FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
filtro5_2
```

```python
#filtro de filas
filtro6_2 = df.iloc[7001:7100,:]
filtro6_2
```

```python
# filtro 7
filtro7_2 = df[(df['CVE_VEND'] == 1) | (df['CVE_VEND'] == 2)] #Únicamente ponemos cve_vend de 1 y 2
filtro7_2 = filtro7_2.set_index('CVE_VEND') #Hacemos la columna cve_vend el índice
filtro7_2 = filtro7_2['FECHAELAB'] #Dejamos como única columna la fecha de elaboración
filtro7_2
```

```python
filtro1_2.to_csv('filtro1_2.csv')
filtro2_2.to_csv('filtro2_2.csv')
filtro3_2.to_csv('filtro3_2.csv')
filtro4_2.to_csv('filtro4_2.csv')
filtro5_2.to_csv('filtro5_2.csv')
filtro6_2.to_csv('filtro6_2.csv')
filtro7_2.to_csv('filtro7_2.csv')
```
